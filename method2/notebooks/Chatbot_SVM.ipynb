Import module

```python
import json
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

Load dataset

```python
# Load dataset
with open('labeled_data.json') as f:
    data = json.load(f)

print(data)
```

Extract prompt, split data, make pipeline and train

```python
# Extract prompts and responses
X = [item['prompt'] for item in data]
y = [item['label'] for item in data]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a pipeline with TF-IDF vectorization and SVM classifier
pipeline = make_pipeline(TfidfVectorizer(), SVC(kernel='linear', C=1.0))

# Train the model
pipeline.fit(X_train, y_train)
```

Evaluation

```python
# Evaluate the model
y_pred = pipeline.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

Test the chatbot

```python
# Function to generate response
def generate_response(prompt):
    return pipeline.predict([prompt])[0]

# Test the chatbot
user_input = "Apa yang kamu lakukan ketika waktu luang?"
response = generate_response(user_input)
print("Chatbot response:", response)
```
